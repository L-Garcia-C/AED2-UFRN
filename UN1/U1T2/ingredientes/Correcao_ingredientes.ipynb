```python
import pandas as pd
```

```python
X=pd.read_csv('ingredientes(erro).csv',sep=';')
X
```

```python
Y=[]
for index, row in X.iterrows():
    linha = row['nome_receita;ingredientes;tipos_ingredientes']
    partes = linha.split(';')
    if len(partes) == 3:
        nome, ingredientes, tipos = partes
        Y.append([nome, ingredientes.strip('"'), tipos.strip('"')])
Y
```

```python
Z=pd.DataFrame(Y,columns=['nome_receita','ingredientes','tipos_ingredientes'])
Z.to_csv('ingredientes.csv')
Z
```
